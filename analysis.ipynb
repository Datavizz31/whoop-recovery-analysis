```python
import pandas as pd

# Load both
raw = pd.read_csv('physiological_cycles.csv')
processed = pd.read_csv('processed_whoop_data.csv')

print("=== RAW (physiological_cycles.csv) ===")
print(f"Rows: {len(raw)}")
print(f"Columns: {len(raw.columns)}")
print(raw.columns.tolist())

print("\n=== PROCESSED (processed_whoop_data.csv) ===")
print(f"Rows: {len(processed)}")
print(f"Columns: {len(processed.columns)}")
print(processed.columns.tolist())

print("\n=== WHAT'S DIFFERENT ===")
raw_cols = set(raw.columns)
processed_cols = set(processed.columns)
print(f"Added in processed: {processed_cols - raw_cols}")
print(f"Removed in processed: {raw_cols - processed_cols}")

print("\n=== EXPERIMENT PERIODS ===")
print(processed['Period'].value_counts())
```

```python
print(processed['Period'].value_counts())
print()
print(processed.groupby('Period')[['Recovery score %', 
    'Heart rate variability (ms)', 
    'Deep sleep %', 
    'REM sleep %']].mean().round(1))
```
